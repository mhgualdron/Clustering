```python
from sklearn.datasets import make_blobs
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from sklearn.cluster import KMeans
```

```python
%run "template_visualitation.ipynb"
```

```python
x,y = make_blobs(n_samples=100, centers=4,n_features=2,
          cluster_std=[1,1.5,2,2],random_state=7)
```

```python
df_blobs = pd.DataFrame({
    'x1':x[:,0],'x2':x[:,1],
    'y':y})
df_blobs
```

```python
def plot2d_clusters(x,y,ax):
    y_uniques = pd.Series(y).unique()
    for i in y_uniques:
        x[y==i].plot(
        title=f'{len(y_uniques)} Clusters',
        kind='scatter',
        x='x1',
        y='x2',
        marker=f'${i}$',
        ax = ax)
```

```python
fig, ax = plt.subplots(1,1, figsize=(9,9))
x,y = df_blobs[['x1','x2']],df_blobs['y']
plot2d_clusters(x,y,ax)
```

```python
kmeans = KMeans(n_clusters=4, random_state=7,n_init=10)
```

```python
y_pred = kmeans.fit_predict(x)
```

```python
fig, ax = plt.subplots(1,2,figsize=(9,9))
plot2d_clusters(x,y,ax[0])
plot2d_clusters(x,y_pred,ax[1])
ax[0].set_title(f'Actual {ax[0].get_title()}')
ax[1].set_title(f'KMeans {ax[1].get_title()}')
```
